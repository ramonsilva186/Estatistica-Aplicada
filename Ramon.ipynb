```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
dados = pd.read_csv('mapa2016_armas_UF8.csv', sep=',')
dados
```

#### Mudando Tipo da coluna 'Taxa Média'

```python
soma = ''
lista = []
for i in dados['Taxa Média']:
    soma = ''
    for j in range(len(i)):
        if i[j] == ',':
          soma += '.'
        else:
            soma += i[j]
    lista.append(float(soma))
```

#### Mudando Tipo da coluna 'Pop. Mé-dia 12/14'

```python
som = ''
lista_2 = []

for i in dados['Pop. Mé-dia 12/14']:
    som = ''
    for j in range(len(i)):
        if i[j] != '.':
            som += i[j]
    lista_2.append(int(som))
```

```python
dados['Taxa Média'] = lista
dados['Pop. Mé-dia 12/14'] = lista_2
dados
```

#### DataFrame do Sudeste

```python
mg = dados[dados['UF'] == 'MG']
es = dados[dados['UF'] == 'ES']
sp = dados[dados['UF'] == 'SP']
rj = dados[dados['UF'] == 'RJ']

sudeste = pd.concat([mg, es, sp, rj])
sudeste
```
