```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
dados = pd.read_csv('mapa2016_armas_UF8.csv', sep=',')
dados
```

#### Mudando Tipo da coluna 'Taxa Média'

```python
soma = ''
lista = []
for i in dados['Taxa Média']:
    soma = ''
    for j in range(len(i)):
        if i[j] == ',':
          soma += '.'
        else:
            soma += i[j]
    lista.append(float(soma))
```

#### Mudando Tipo da coluna 'Pop. Mé-dia 12/14'

```python
som = ''
lista_2 = []

for i in dados['Pop. Mé-dia 12/14']:
    som = ''
    for j in range(len(i)):
        if i[j] != '.':
            som += i[j]
    lista_2.append(int(som))
```

```python
dados['Taxa Média'] = lista
dados['Pop. Mé-dia 12/14'] = lista_2
dados
```

#### DataFrame de cada estado da região Sudeste

```python
mg = dados[dados['UF'] == 'MG']
es = dados[dados['UF'] == 'ES']
sp = dados[dados['UF'] == 'SP']
rj = dados[dados['UF'] == 'RJ']
```

#### Ordenando as colunas '2012' , '2013', '2014' de cada estado 

```python
mg = mg.sort_values(['2012', '2013', '2014*'])
es = es.sort_values(['2012', '2013', '2014*'])
sp = sp.sort_values(['2012', '2013', '2014*'])
rj = rj.sort_values(['2012', '2013', '2014*'])
sudeste = pd.concat([mg, es, sp, rj])
arr = np.arange(893)
sudeste = sudeste.set_index(arr)
```

#### DataFrame com o maior de cada estado da região

```python
maior_se = pd.concat([df6, df7, df8, df9])
```

#### Função ajusta o dataframe

```python
def teste(a):
    a.drop('UF', axis= 1, inplace=True)
    a.drop('Pos. Nac.', axis= 1, inplace=True)
    a.drop('Taxa Média', axis= 1, inplace=True)
    a.drop('Pop. Mé-dia 12/14', axis= 1, inplace=True)
    
    inde = []
    dados_12 = []
    dados_13 = []
    dados_14 = []
    for i in a['2012']:
        dados_12.append(i)
    for i in a['2013']:
        dados_13.append(i)
    for i in a['2014*']:
        dados_14.append(i)
    for i in a['Município']:
        inde.append(i)

    df5 = pd.DataFrame({
                       '2012' : dados_12,
                       '2013' : dados_13,
                       '2014' : dados_14}, index=inde)
    return df5
```

```python
a = teste(maior_se)
a
```

```python
a.plot.bar()
plt.grid(ls=':')
plt.title("Sudeste - HAF")
plt.xlabel("Cidades")
plt.ylabel("Quantidade de Haf")
```
